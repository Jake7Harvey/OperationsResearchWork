$$
\sum_{i = 0}^n i^3 = \left(\frac{n(n+1)}{2}\right)^2
$$

```python
from sympy import *           #Code without true or false

def SumOfCubes(x):
  SumOfCubesAnswer = 0
  for num in range(x + 1):
    SumOfCubesAnswer = SumOfCubesAnswer + num**3
  return SumOfCubesAnswer

```
